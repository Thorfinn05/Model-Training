```python
from ultralytics import YOLO

# load YOLOv8 small model (better than nano, still light)
model = YOLO("yolov8s.pt")

# train on your dataset
model.train(
    data="/content/dataset/data.yaml",  # path to your yaml
    epochs=50,
    imgsz=640
)
```

```python
# load best model
model = YOLO("runs/detect/train3/weights/best.pt")

# run prediction on one image
results = model.predict(
    source="/content/dataset/train/images/10_jpg.rf.6aaa9b4b59c22962b72daa6f7c782594.jpg",
    show=True,
    save=True
)
```
